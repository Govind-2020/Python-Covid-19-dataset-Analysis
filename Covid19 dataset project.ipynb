```python
#18.How does the fatality rate vary across different provinces/states?
plt.figure(figsize=(12, 8))
df['State']= df['Province/State'].iloc[:10]
df['FatalityRate'] = (df['Deaths'] / df['Confirmed']) * 100 
sns.boxplot(x='State', y='FatalityRate', data=df)
```

```python
#20.What is the trend of active cases over time globally?
# Calculate active cases
df['ActiveCases'] = df['Confirmed'] - df['Recovered'] - df['Deaths']
df['ActiveCases'].iloc[:50]
plt.figure(figsize=(12, 8))
sns.lineplot(x='Observation Date', y='ActiveCases', data=df, marker='o')

plt.title('Trend of Active Cases Over Time Globally')
plt.xlabel('Date')
plt.ylabel('Active Cases')
plt.show()
```
